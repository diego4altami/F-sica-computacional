```python
g, k = 9.81, 0.5
```

```python
y0, v0 = 10, 100
z0 = np.array([y0, v0])
```

```python
#tiempo de vuelo
tv = (v0+np.sqrt(v0**2+2*g*y0))/g
```

```python
tInc = 1/q
tStop = (1.+ tInc)*tv
t = np.arange(0., tStop, tInc)
```

```python
def Euler(f_user, U0, t, *params):

    n = t.size
    
    neq = U0.size
    u = np.zeros((n, neq))
    
    u[0] = U0
    dt = (t[1]-t[0])/q
        
    for k in range(n-1):
        u[k+1] = u[k] + dt*f_user(u[k], t[k],*params)
        
    return u
```

```python
def Heun(f_user, U0, t, *params):

    n  = t.size
    neq = U0.size
    u = np.zeros((n, neq))

    u[0] = U0
    h = (t[1]-t[0])/q
    
    for i in range(n-1):
        k1 = f_user(u[i], t[i],*params)
        k2 = f_user(u[i]+k1*h, t[i+1],*params)
        u[i+1] = u[i] + h* (k1+k2)/2.
    return u
```

```python
def Runge_kutta_4to(f_user, U0, t, *params):

    n = t.size
    neq = U0.size
    u = np.zeros((n, neq))

    u[0] = U0
    h = (t[1]-t[0])/q

    for i in range(n-1):
        k1 = f_user(u[i], t[i],*params)
        k2 = f_user(u[i]+(1/2)*k1*h, t[i]+(1/2)*h, *params)
        k3 = f_user(u[i]+(1/2)*k2*h, t[i]+(1/2)*h, *params)
        k4 = f_user(u[i]+k3*h, t[i]+h, *params)
        u[i+1] = u[i] + h*(k1+2*k2+2*k3+k4)/6.
    return u
```
