```python
#posición inicial
x0=0
y0=0
#ángulo inicial
theta0=45
#velocidad inicial y sus componentes
v0=10
vx0=v0*np.cos(theta0)
vy0=v0*np.sin(theta0)
#parametros del problema
n=1
k=0
g=9.81
#vector inicial
u0 = np.array([x0,vx0,y0,vy0])
```

```python
#función que describe el movimiento de in proyectil
def Proyectil(u,n,k,g):
    rapid=np.sqrt(u[1]**2+u[3]**2)
    derivs = np.array(u[1],-k*rapid**(n-1),u[3], -g-k*rapid**(n-1))
    return derivs
```

```python
def Runge_kutta_4to(f_user, U0, t, *params):

    n = t.size
    neq = U0.size
    u = np.zeros((n, neq))

    u[0] = U0
    h = (t[1]-t[0])/q

    for i in range(n-1):
        print(u[i])
        print(t[i])
        print(*params)
        print(f_user(u[i], t[i],*params))
        k1 = f_user(u[i], t[i],*params)
        print(k1)
        k2 = f_user(u[i]+((1/2)*k1*h), t[i]+((1/2)*h), *params)
        k3 = f_user(u[i]+((1/2)*k2*h), t[i]+((1/2)*h), *params)
        k4 = f_user(u[i]+k3*h, t[i]+h, *params)
        u[i+1] = u[i] + h*(k1+2*k2+2*k3+k4)/6.
    return u
```

```python
proyect=Runge_kutta_4to(Proyectil,u0,20.0,n,g,k)
```
