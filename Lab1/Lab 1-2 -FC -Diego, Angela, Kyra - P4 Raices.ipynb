```python
#Definimos una función cualquiera para sacarle las raíces

def funcion(x):
    return ((x+2)*(x-3)*(x+8.5))


#Graficamos la función para poder ubicar las raíces 

import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.arange(-10,10,0.1),funcion(np.arange(-10,10,0.1)))
plt.grid()
```

```python
#Obtiene la raíz con el método de bisección con un error de 0.0001
def bis(x1,x2,funcion):
    pm = 1
    error=0.0001
    while abs(funcion(pm))>error:
        if (funcion(x1)*funcion(x2))<0:
            pm = (x1+x2)/2
            if (funcion(pm)*funcion(x2))<0:
                x1 = pm
            else:
                x2 = pm
        else:
            pm = (x1+x2)/2
            x1 = pm
    return pm
                
        
```

```python
def buscarraiz(a,b,funcion):
    
    raices=[]
    for i in range(a,b,1):
        
        #Raices enteras
        if funcion(i+1)==0:
            if (funcion(i)*funcion(i+1))<0:
                print("raíz: ", bis(i,i+1,funcion))
                raices.append(bis(i,i+1,funcion))
            
            if (funcion(i)*funcion(i+1))==0:
                print("raíz: ", bis(i,i+1,funcion))
                raices.append(bis(i,i+1,funcion))
        #Raices decimales
       
        elif (funcion(i)*funcion(i+2))<0:
            print("raíz: ", bis(i,i+2,funcion))
            raices.append(bis(i,i+2,funcion))
            
        elif (funcion(i)*funcion(i+2))==0:
            print("raíz : ", bis(i,i+2,funcion))
            raices.append(bis(i,i+2,funcion))
        
   
    return(raices)     

        
        
```
