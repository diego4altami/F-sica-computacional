# Pregunta 3, Laboratorio 1 y 2: Derivada centrada

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def centrada(f,a,b,N):
    h = (b-a)/N
    d = (f(np.arange(a+h,b,h))-f(np.arange(a-h,b+a*h,h)))/(2*h)
    return d
```

```python
def funcion1(x):
    return 1 + 1/2*np.tanh(2*x)
```

```python
a=-2
b=2
N=100
d = centrada(funcion1,a,b,N)
print(d)
```

Gráfica de la función

```python
def derivada_exacta(x):
    return ((1/(np.cosh(2*x)))**2)

derivada_exacta(2)
```

```python
N1=5
N2=10
N3=20
N4=1000
h1=(b-a)/N1
h2=(b-a)/N2
h3=(b-a)/N3
h4=(b-a)/N4


plt.plot(np.arange(a,b-h1,h1),centrada(funcion1, a,b,N1))
plt.plot(np.arange(a,b-h2,h2),centrada(funcion1, a,b,N2))
plt.plot(np.arange(a,b-h3,h3),centrada(funcion1, a,b,N3))
plt.plot(np.arange(a,b-h4,h4),centrada(funcion1, a,b,N4))
plt.xlabel('x')
plt.ylabel('y')
plt.title('derivada (programada) función 1')
plt.show()

```

Conforme aumentamos el valor de N nuestra derivada se parece más a la derivada exacta. Aumentar el valor de N no significa aumentar la exactitud de la derivada si no aumentar los intervalos en  los cuales se va a calcular. 


La derivada por naturaleza va a contener un error dado por lo términos no calculados de la serie de Taylor

```python
def exacta(Na):
    h=(b-a)/Na
    x=np.arange(a,b-h,h)
    return(derivada_exacta(x))
h=(b-a)/N
x=np.arange(a,b-h,h)

plt.plot(x,exacta(100))
plt.xlabel('x')
plt.ylabel('d(x)')
plt.title('derivada (exacta) función 1')
plt.show()
```

Obtención del error para n=1000

```python
valor_real= exacta(1000)
valor_aproximado= centrada(funcion1, a,b,N4)
error= abs(valor_real-valor_aproximado)
print(np.mean(error)) #Se tomó el valor medio del error al comparar cada punto con la derivada exacta en ese punto

```

```python
#Gráfica del error para N=1000

plt.plot(np.arange(a,b-h4,h4),error,color='red')
plt.xlabel('x')
plt.ylabel('error')
plt.title('Error función 1')
plt.show()
```
