Definimos la función para obtener sus raices 

```python
import math
```

```python
def funcion(x):
    return ((5*(1/(math.e**(x))))+x-5)
```

Graficamos la función para poder ubicar las raíces

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.arange(-1,10,0.1),funcion(np.arange(-1,10,0.1)))
plt.grid()
```

```python
def biseccion(f,x1,x2):
    pm=0
    i=0
    while(f(x1)*f(x2))<0:
        pm = (x1+x2)/2
        if f(pm)*f(x1)<0:
            x2 = pm 
            i += 1
        else:   
            x1 = pm
            i += 1
        if i > 10000:
            break 
    return pm
```

```python
def buscarraiz(a,b,funcion): 
    for i in range(a,b,1):
        if (funcion(i)*funcion(i+1))<=0:
            print("raíz: ", biseccion(funcion,i,i+1))
        elif (funcion(i)*funcion(i+2))<=0:
            print("raíz: ", biseccion(funcion,i,i+2))
```

```python
a = -10
b = 10
buscarraiz(a,b,funcion)
```
