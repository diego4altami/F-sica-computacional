# MÃºltiples muestras para una red cuadrada

```python
import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt
```

```python
def permutation():
    global order,N
    order = nr.permutation(N)

# Setting the boundaries no PBC utilizando case, no hay vecino EMPTY
def vecinocuad(s,j):
    global L,empty

    if j == 0: #izquierda
        if s%L == L-1:
            return empty
        else:
            return s+1
        
    elif j==1: #arriba
        if s//L == L-1:
            return empty
        else:
            return s+L
        
    elif j==2: #derecha
        if s%L == 0:
            return empty
        else:
            return s-1
        
    elif j==3: #abajo
        if s//L == 0:
            return empty
        else:
            return s-L
                
def boundaries():
    global N,nn
    for s in range(N):
        for j in range(4):
            nn[s,j] = vecinocuad(s,j)

# Find operation de la raiz, versiÃ³n no recursiva, no se cambian los ptr.
def findroot(r):
    global parent
    while parent[r] >= 0:
        r = parent[r]
    return r
```

```python
#union de raÃ­ces (clusters), version no recursiva

    
def mergeroots(r1,r2):
    global spanclussize,parent,touchesLeft,touchesRight
    if r1 == r2:
        return r1
    elif -parent[r1] > -parent[r2]:
        parent[r1] += parent[r2]
        parent[r2] = r1
        touchesLeft[r1] = touchesLeft[r1] or touchesLeft[r2] 
        touchesRight[r1] = touchesRight[r1] or touchesRight[r2]
        if touchesLeft[r1] and touchesRight[r1]:
            spanclussize = -parent[r1]
        return r1 
    else:
        parent[r2] += parent[r1]
        parent[r1] = r2
        touchesLeft[r2] = touchesLeft[r2] or touchesLeft[r1]
        touchesRight[r2] = touchesRight[r2] or touchesRight[r1]
        if touchesLeft[r2] and touchesRight[r2]:
            spanclussize = -parent[r2]
        return r2  
```

```python
def percolate():
    global spanclussize,order,parent,nn
    BIG = []
    PSpan = []
    big=0
    spanclussize = 0
    
    for i in range(N): parent[i]=empty    

    for i in range(N):
        r1=s1=order[i]
        parent[s1] = -1
                
        for j in range(4):
            s2=nn[s1, j]
            if s2 != empty:
                if parent[s2] != empty:
                    r1 = mergeroots(r1,findroot(s2))
                    if -parent[r1]>big: big=-parent[r1]
        BIG.append(big)
        PSpan.append(spanclussize/(i+1.))

    return np.array(BIG), np.array(PSpan)
```

```python
#Datos, inicalizaciones  y corridas

global L, N, empty, parent, nn, order, spanclussize,touchesLeft,touchesRight

L = 128   #no. de celdas por lado
N = L**2   #no. de celdas
empty=-(N+1)
muestras=100 #no. de realizaciones

bigprom=np.zeros(N)  #promedio del tamaÃ±o de los arreglos 
Pspanprom=np.zeros(N) #prob de que el sistema percole promedio
Probprom=np.zeros(N)   #probabilidad de ocupaciÃ³n promedio
nn=np.zeros((N, 4), dtype=int)
boundaries() 

for i in range (muestras):
    order=np.zeros(N, dtype=int) 
    parent=np.zeros(N, dtype=int) 
    touchesLeft =  [i%L == 0 for i in range(N)]
    touchesRight = [i%L == L-1 for i in range(N)]

    permutation()  
    BIG,PSpan=percolate()
    Prob = np.copy(PSpan)
    Prob[PSpan>0]=1

    bigprom+=BIG
    Pspanprom+=PSpan
    Probprom+=Prob

bigprom/=muestras     #promedio del tamaÃ±o del cluster mÃ¡s grande
Pspanprom/=muestras  #prob de que una celda ocupada pertenezca al cluster que percola 
Probprom/=muestras  #prob promedio de percolaciÃ³n 
```

```python
#GrÃ¡ficas

#1) GrÃ¡fica de p vs. promedio en el tamaÃ±o del cluster mÃ¡s grande

p = np.array(range(1,N+1))/N #se divide entre n para normalizar
plt.figure(figsize=(8,6))
bigprom/=N #para normalizar 
plt.plot(p, bigprom, color="red",label = "Largest cluster")
plt.xlabel(r"$p$",fontsize = 18)
plt.ylabel("Average largest cluster size",fontsize = 18)
plt.title(r"$N = $ {} $\times$ {}".format(L,L))
plt.axvline(0.59275,color = "gray",zorder=-1)
plt.text(0.6,0.9*bigprom.max(),r'$p_c = 0.59275$')
plt.legend(loc = 'best')
plt.show()

#2) GrÃ¡fica de p vs. probabilidad promedio de que una celda ocupada pertenezca al cluster que percola


plt.figure(figsize=(8,6))
plt.plot(p, Pspanprom, color="purple",label = "Spanning cluster size/ Occupation size")
plt.xlabel(r"$p$",fontsize = 24)
plt.ylabel(r'$Average \ P_\infty(p)$',fontsize = 24)
plt.title("ParÃ¡metro de Orden")
plt.axvline(0.59275,color = "gray",zorder=-1)
plt.text(0.35,0.9,r'$p_c = 0.59275$')
plt.legend()
plt.show()

#3) GrÃ¡fica de p vs. probabilidad promedio de percolaciÃ³n 


plt.figure(figsize=(8,6))
plt.plot(p, Probprom, 'b-',label = "percolation probability")
plt.xlabel(r"$p$",fontsize = 24)
plt.ylabel(r'$Average \ prob \ of \ percolation$',fontsize = 24)
plt.title("Probabilidad de percolaciÃ³n")
plt.axvline(0.59275,color = "gray",zorder=-1)
plt.text(0.35,0.9,r'$p_c = 0.59275$')
plt.legend()
plt.show()
```

# Varios tamaÃ±os del sistema

```python
global L, N, empty, parent, nn, order, spanclussize,touchesLeft,touchesRight

#INTENATANDO USAR UN FORRRRR
lados = [16,32,64,128,256]   #no. de celdas por lado
bigprom_L={}  #diccionario vacÃ­o
Pspanprom_L={}
Probprom_L={}
    
for x in range (0,5,1):
    L=lados[x]
    N=L**2
    empty=-(N+1)
    muestras=100
    bigprom=np.zeros(N)  #promedio del tamaÃ±o de los arreglos 
    Pspanprom=np.zeros(N) #prob de que el sistema percole promedio
    Probprom=np.zeros(N)   #probabilidad de ocupaciÃ³n promedio
    nn=np.zeros((N, 4), dtype=int)
    boundaries() 

    for i in range (muestras):
        order=np.zeros(N, dtype=int) 
        parent=np.zeros(N, dtype=int) 
        touchesLeft =  [i%L == 0 for i in range(N)]
        touchesRight = [i%L == L-1 for i in range(N)]

        permutation()  
        BIG,PSpan=percolate()
        Prob = np.copy(PSpan)
        Prob[PSpan>0]=1

        bigprom+=BIG
        Pspanprom+=PSpan
        Probprom+=Prob

    bigprom/=muestras     #promedio del tamaÃ±o del cluster mÃ¡s grande
    Pspanprom/=muestras  #prob de que una celda ocupada pertenezca al cluster que percola 
    Probprom/=muestras  #prob promedio de percolaciÃ³n 
        
    bigprom_L[x]=bigprom
    Pspanprom_L[x]=Pspanprom
    Probprom_L[x]=Probprom

print("El arreglo mÃ¡s grande promedio como funciÃ³ de p es: ",bigprom)
print("El Pspan promedio como funciÃ³ de p es: ",Pspanprom)
print("La probabilidad de que percole promedio como funciÃ³ de p es: ",Probprom)
```

```python

#GrÃ¡ficas para diferentes tamaÃ±os de L


#1) GrÃ¡fica de p vs. probabilidad promedio de que una celda ocupada pertenezca al cluster que percola

for x in range (0,5,1):
    p = np.array(range(1,lados[x]**2+1))/(lados[x]**2)
    plt.plot(p, Pspanprom_L[x], label = "Spanning cluster")
plt.xlabel(r"$p$",fontsize = 24)
plt.ylabel(r'$Average \ P_\infty(p)$',fontsize = 24)
plt.title("ParÃ¡metro de Orden")
plt.axvline(0.59275,color = "gray",zorder=-1)
plt.legend()
plt.show()

#2) GrÃ¡fica de p vs. probabilidad promedio de percolaciÃ³n 

for x in range(0,5,1):
    p = np.array(range(1,lados[x]**2+1))/(lados[x]**2)
    plt.plot(p, Probprom_L[x], label = "percolation probability")
plt.xlabel(r"$p$",fontsize = 24)
plt.ylabel(r'$Average \ prob \ of \ percolation$',fontsize = 24)
plt.title("Probabilidad de percolaciÃ³n")
plt.axvline(0.59275,color = "gray",zorder=-1)
plt.legend()
plt.show()
```

a) La mÃ¡s ruidosa es la que tiene menos lados y mientras mÃ¡s casillas hay en el sistema se vuelve mÃ¡s suave. Conforme se aumenta el nÃºmero de lados, las desviaciones se reducen, por lo que cuando L tiende a infinito, las desviaciones tienden a cero. Esto de debe a que la ocupaciÃ³n es un proceso aleatorio por lo que mientras menos casillas tengamos mÃ¡s probabilidad hay de que se aparten de la tendencia. 

b) Para cuando L tiende a infinito, la grÃ¡fica de parÃ¡metro de orden se volverÃ­a muy suave, y la transiciÃ³n de 0 a 1 serÃ­a muy pronunciada. La grÃ¡fica de probabilidad de percolaciÃ³n serÃ­a un escalÃ³n: 0 antes de p_c y 1 despuÃ©s de p_c.
