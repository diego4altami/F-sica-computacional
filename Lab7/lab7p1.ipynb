```python
from random import random,randrange,seed
import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt
```

```python
def GasIdealClasicod1(T,v0,vmax,steps):
    beta = 1./T
    k0 = 0.5*v0*v0
    energia = [k0]
    Velocity = [v0]

    for mcs in range(steps):
        phi=random()*2*np.pi
        vprueba = v0  + random()*vmax*(np.cos(phi)+np.sin(phi))
        kprueba = 0.5*vprueba*vprueba
        dk = kprueba-k0
        if dk < 0 or random() < np.exp(-beta*dk):
            v0 = vprueba
            k0 = kprueba
        energia.append(k0)
        Velocity.append(v0)
    return np.array(energia),np.array(Velocity)
```

```python
T = 10
v0 = 10 
vmax= 5 
steps = 1000000
energia,velocidad = GasIdealClasicod1(T,v0,vmax,steps)
```
