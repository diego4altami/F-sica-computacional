```python
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
#la curva de koch

def curvaKoch(x1,y1,x2,y2,n):
    '''
        x1,y1: punto izquierdo de la línea a subdividir
        x2,y2: punto derecho de la línea a subdividir
        n: número de interaciones
    '''
    if n>0:
        dx = (x2-x1)
        dy = (y2-y1)
        x1p = x1+dx/3
        y1p = y1+dy/3
        x2p = x1+2*dx/3
        y2p = y1+2*dy/3
        x3p = x1p + dx/6 - np.sqrt(3)*dy/6
        y3p = y1p + dy/6 + np.sqrt(3)*dx/6
        
        curvaKoch(x1,y1,x1p,y1p,n-1)
        curvaKoch(x1p,y1p,x3p,y3p,n-1)
        curvaKoch(x3p,y3p,x2p,y2p,n-1)
        curvaKoch(x2p,y2p,x2,y2,n-1)
    else:
        plt.plot([x1,x2],[y1,y2],'b',linewidth=2.0)
```

```python
fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim(-300,300)
ax.set_ylim(-2,300*3/4-2)
ax.set_xticks([])
ax.set_yticks([])
curvaKoch(-300,0,300,0,2)
```

```python
import matplotlib.pyplot as plt

def curvaKochcuad(x1, y1, x2, y2, n):
    '''
        x1, y1: punto izquierdo de la línea a subdividir
        x2, y2: punto derecho de la línea a subdividir
        n: número de iteraciones
    '''
    if n > 0:
        dx = x2 - x1
        dy = y2 - y1
        x1p = x1 + dx / 3
        y1p = y1 + dy / 3 
        x4p = x1p +  dx / 3
        y4p = y1p + dy / 3
        x2p = x1p + dx / 3
        y2p = y1p + (dy + dx) / 3
        x3p = x4p - (dx - dy) / 3
        y3p = y4p - (dy + dx) / 3


        curvaKochcuad(x1p, y1p, x2p, y2p, n - 1)
        curvaKochcuad(x2p, y2p, x3p, y3p, n - 1)
        curvaKochcuad(x3p, y3p, x4p, y4p, n - 1)
    else:
        plt.plot([x1, x2], [y1, y2], 'b', linewidth=2.0)

fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim(-300,300)
ax.set_ylim(-2,300*3/4-2)
ax.set_xticks([])
ax.set_yticks([])
curvaKochcuad(-300, 0, 300, 0, 2)
```
