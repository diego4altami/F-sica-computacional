(b) Modifique su programa para simular el lanzamiento de dos dados N = 1000000 veces y cuente el nu ́mero de veces que obtenga doble seis. Divida entre N para que obtenga la fraccio ́n de veces que usted obtenga doble seis

```python
def dado(N):
    cont = 0
    for i in range(N):
        dado1=nr.choice([1,2,3,4,5,6])
        dado2=nr.choice([1,2,3,4,5,6])
        if dado1+dado2==12:
            cont += 1
    frac = cont/N
    return frac
```

```python
prueba1 = dado(1000000)
```

```python
real = 1/36
print(real)
```

(c) Calcule el error porcentual de su estimado. ¿Qu ́e pasar ́a con este error si N = 10000?

```python
error1 = abs(prueba1-real)/real*100
print(error1)
```

```python
prueba2 = dado(10000)
```

```python
error2 = abs(prueba2-real)/real*100
print(error2)
```

```python
deltaerror=error2-error1
print(deltaerror)
```
